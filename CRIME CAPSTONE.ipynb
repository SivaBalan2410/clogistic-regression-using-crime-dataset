```python
crimes = data.groupby('STATE/UT').sum()
```

```python
def setHighCrime(df2):
    '''Function to set value of highCrime depending on ViolentCrimesPerPop'''
    if df2['DOWRY DEATHS'] > 7:
        return True
    else:
        return False
    
# Adding a new field "highCrime"
data['highCrime'] = data.apply(setHighCrime, axis=1)

# Calculating the percentage of positive and negative instances in the dataset
percentage_intances = data.groupby('highCrime').size() * 100 / len(data)
print(percentage_intances)
```

```python
X = data.drop('RAPE', axis=1).drop('IMPORTATION OF GIRLS', axis=1).drop('INSULT TO MODESTY OF WOMEN', axis=1).drop('CRUELTY BY HUSBAND OR HIS RELATIVES', axis=1).drop('STATE/UT', axis=1).drop('DOWRY DEATHS', axis=1).drop('DISTRICT', axis=1).drop('ASSAULT ON WOMEN WITH INTENT TO OUTRAGE HER MODESTY', axis=1).drop('KIDNAPPING AND ABDUCTION', axis=1).drop('highCrime', axis=1)

y = data["highCrime"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,y,test_size = .50,random_state = 10)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,precision_recall_curve, roc_curve, auc, log_loss
import numpy as np
logreg = LogisticRegression()
logreg.fit(x_train,y_train)

y_pred = logreg.predict(x_test)
logreg_accy = round(accuracy_score(y_pred,y_test), 3)
print (logreg_accy)
```

```python
print (classification_report(y_test, y_pred, labels=logreg.classes_))
print (confusion_matrix(y_pred, y_test))
```
